# GeoCoding to get Chinatwon in the world 

```python
import geopy
import pandas as pd
from geopy.geocoders import MapBox
import geocoder
import numpy as np
import geopandas
```

```python
# load the data 
data = pd.read_csv('data/chinatown.csv')
```

```python
data.head()
```

```python
from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim
longitude = []
latitude = []
def findGeocode(city): 
    try:
        geolocator = Nominatim(user_agent="what_ever_just_run")
        return geolocator.geocode(city)
    except GeocoderTimedOut:
        return findGeocode(city)    
  
for i in (data["city"]):
    if findGeocode(i) != None:
        loc = findGeocode(i)
        latitude.append(loc.latitude)
        longitude.append(loc.longitude)
    else:
        latitude.append(np.nan)
        longitude.append(np.nan)
```

```python
data["Longitude"] = longitude
data["Latitude"] = latitude
data
```

```python
#save as csv 
data.to_csv('chinatown_geo.csv', encoding='utf-8')
```

```python
#save to geojson 
gdf = geopandas.GeoDataFrame(
    data, geometry=geopandas.points_from_xy(data.Longitude, data.Latitude))
```

```python
gdf.head()
```

```python
with open('chinatown.geojson' , 'w') as file:
    file.write(gdf.to_json())
```
